# Correlation Analysis

```python
nutrients_values=['calories_kcal', 'fats_g', 'sod_mg', 'carb_g', 'fiber_g',
       'sugar_g', 'protein_g', 'vitA_g', 'calcium_mg', 'thiamin_mg', 'zinc_mg',
       'potassium_mg', 'magnesium_mg', 'vitE_mg', 'vitK_mcg', 'vitC_mg',
       'vitB6_mg', 'copper_mg', 'carotene_mg', 'carotene_mcg',
       'cryptoxanthin_mcg', 'lycopene_mcg', 'cholesterol_mg','quality']

#calculating correlation Matrix Among Nutrient Value
corr_matrix = df_cleaned[nutrients_values].corr().round(2)
print(corr_matrix)
```

```python
# Creating a heatmap of the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix among Nutrients')
plt.show()
```

Using a heatmap to visualize correlations in order to provide a clear and intuitive understanding of how different nutrients relate to each other.

**positive correlation:** "vitA_g" and "carotene_mg" show a high positive correlation, it indicates that foods high in calcium are also likely to be high in protein.

**negative correlation:** "fat_g" and "vitC_mg" have a negative correlation, it suggests that foods high in sodium tend to be lower in potassium.

```python
# Calculate the correlation between quality and nutrient columns
correlation_matrix = df_cleaned[nutrients_values].corr().round(2)

# Extract the correlation values with quality
quality_correlations = correlation_matrix['quality'].drop('quality')

# Convert to DataFrame for better visualization
quality_correlations_df = quality_correlations.to_frame()

print(quality_correlations)
```

```python

plt.figure(figsize=(12, 10))
sns.heatmap(quality_correlations_df, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix between quality and Nutrients')
plt.show()
```

Understanding these correlations helps in identifying which nutrients most strongly influence the quality. For example,
The "sugar_g" attribute has the strongest negative correlation with quality (-0.79). This suggests that as the sugar content increases, the quality tends to decrease significantly. Hence, reducing sugar content might significantly improve quality

# Nutritional Analysis


```python
variance_values = df_cleaned[nutrients_values].var()
print(variance_values)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=variance_values.index, y=variance_values.values, palette='coolwarm')
plt.xticks(rotation=90)
plt.title('Variance of Nutrient Columns')
plt.xlabel('Nutrient')
plt.ylabel('Variance')
plt.show()
```

Understanding the variance in dataset can help identify which nutrients have the most variability, and therefore may have a bigger impact on the overall nutritional profile of the foods.

**High Variance:** Indicates that the values of that nutrient vary widely across different food items. For example, "vitA_g" and "carotene_mg" have very high variances, indicating significant differences in these nutrients among the foods in the dataset.

**Low Variance:** Indicates that the values are more consistent and close to the mean. For example, "thiamin_mg", "copper_mg", and "vitB6_mg" have very low variances, suggesting that most foods in the dataset have similar amounts of these nutrients.
